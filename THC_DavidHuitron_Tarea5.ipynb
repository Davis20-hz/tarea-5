```python
def acceso_boveda(edad_juan, edad_pedro, num_nietos):
    # Condición a
    condicion_a = (edad_juan > 35) or (edad_pedro < 50)
    # Condición b
    condicion_b = (edad_juan + edad_pedro) % 2 == 0

    # Condición c
    condicion_c = num_nietos >= 1

    # Evaluar todas las condiciones
    if condicion_a and condicion_b and condicion_c:
        return "SI"
    else:
        return "NO"



# Ejemplo de uso:
juan = int(input("Ingrese la edad de Juan: "))
pedro = int(input("Ingrese la edad de Pedro: "))
nietos = int(input("Ingrese el número de nietos de Manuel: "))

Par= (juan + pedro) % 2 == 0
print("la suma de sus edades es par:", Par)

resultado = acceso_boveda(juan, pedro, nietos)
print("¿Tienen acceso a la bóveda?:", resultado)
```

```python
# Ejercicio 2: Reloj mundial usando listas, diccionarios y funciones lambda

# Lista de ciudades con sus diferencias horarias respecto a Ciudad de México (en horas)
ciudades = [
    ("Tokio, Japón", 15),
    ("Pekín, China", 14),
    ("Dusambé, Tayikistán", 11),
    ("Perth, Australia", 14),
    ("Katmandú, Nepal", 11.75),
    ("Vilna, Lituania", 8),
    ("Freetown, Sierra Leona", 6),
    ("Río de Janeiro, Brasil", 3),
    ("Tegucigalpa, Honduras", 0),
    ("La Paz, Baja California Sur, México", -1)
]

# Función lambda para calcular la hora local
calcular_hora = lambda hora_mex, diferencia: (hora_mex + diferencia) % 24

# Hora en Ciudad de México
hora_mex = float(input("Ingrese la hora en Ciudad de México (en formato 24h): "))

# Mostrar resultados
print("Hora en diferentes ciudades del mundo")
for ciudad, diferencia in ciudades:
    hora_local = calcular_hora(hora_mex, diferencia)
    hora_int = int(hora_local)
    minutos = int((hora_local - hora_int) * 60)
    print(f"{ciudad:<40} = {hora_int:02d}:{minutos:02d}")
```

```python
def es_primo():
  while True:
    try:
      n = int(input("Ingrese un número entero: "))
      if n <= 1:
        print("El número debe ser mayor que 1.")
        continue
      Primo=True
      for i in range(2, n):
        if n % i == 0:
          Primo=False
          break
      if Primo:
        print(f"{n} es un número primo.")
        break
    except Error:
      print("Entrada inválida. Por favor, ingrese un número entero.")

es_primo() # Call the function to execute the code
```

```python
def analizar_numero():
  while True:
    try:
      N = int(input("Ingrese un número entero mayor que 100: "))
      if N <= 100:
        print("Error en el número. Intente nuevamente.")
        continue
      # Get the last digit of N
      u = N % 10
      if u in [4, 7, 9]:
        print(f"{N} es divisible por 3:", N % 3 == 0)
      elif u in [1, 5]:
        print("Resultado de dividir entre 10:", N // 10)
      else:
        print("Residuo de dividir entre 3:", N % 3)
      break # Exit the loop after successful input and analysis
    except ValueError:
      print("Por favor, ingrese un número válido.")

analizar_numero()
```

```python
import math

def mcd_mcm():
    try:
        a = int(input("Ingrese el primer número: "))
        b = int(input("Ingrese el segundo número: "))

        mcd = math.gcd(a, b)
        mcm = abs(a * b) // mcd

        resultados = {
            "Número A": a,
            "Número B": b,
            "MCD": mcd,
            "MCM": mcm
        }

        print("Resultados")
        for clave, valor in resultados.items():
            print(f"{clave}: {valor}")

    except Error:
        print("Error: ingrese solo números enteros válidos.")

mcd_mcm()
```

```python
def filtrar_datos_personalizado():
    print("Ingrese los elementos de su lista separados por comas (,)")
    entrada = input("Ejemplo: 25, True, 'texto', 3.14, False, 42: ")

    # Convertimos los datos de texto en valores Python reales
    # eval() interpreta True, False, números y cadenas correctamente
    try:
        lista = [eval(x.strip()) for x in entrada.split(',')]
    except Exception:
        print("Error: asegúrese de escribir correctamente los valores (True, False, números o 'texto').")
        return

    enteros, flotantes, cadenas, booleanos = [], [], [], []

    for d in lista:
        if isinstance(d, bool):
            booleanos.append(d)
        elif isinstance(d, int):
            enteros.append(d)
        elif isinstance(d, float):
            flotantes.append(d)
        elif isinstance(d, str):
            cadenas.append(d)

    print("Resultados del filtrado ")
    print("Enteros:", enteros)
    print("Flotantes:", flotantes)
    print("Cadenas:", cadenas)
    print("Booleanos:", booleanos)

filtrar_datos_personalizado()
```
