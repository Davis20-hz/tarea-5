```python
def acceso_boveda(edad_juan, edad_pedro, num_nietos):
    # Condición a
    condicion_a = (edad_juan > 35) or (edad_pedro < 50)
    # Condición b
    condicion_b = (edad_juan + edad_pedro) % 2 == 0

    # Condición c
    condicion_c = num_nietos >= 1

    if condicion_a and condicion_b and condicion_c:
        return "SI"
    else:
        return "NO"

juan = int(input("Ingrese la edad de Juan: "))
pedro = int(input("Ingrese la edad de Pedro: "))
nietos = int(input("Ingrese el número de nietos de Manuel: "))

Par= (juan + pedro) % 2 == 0
print("la suma de sus edades es par:", Par)

resultado = acceso_boveda(juan, pedro, nietos)
print("¿Tienen acceso a la bóveda?:", resultado)
```

```python
# Ciudades con sus diferencia en horas respecto a Ciudad de México
ciudades = [
    ("Tokio, Japón", 15),
    ("Pekín, China", 14),
    ("Dusambé, Tayikistán", 11),
    ("Perth, Australia", 14),
    ("Katmandú, Nepal", 11.75),
    ("Vilna, Lituania", 8),
    ("Freetown, Sierra Leona", 6),
    ("Río de Janeiro, Brasil", 3),
    ("Tegucigalpa, Honduras", 0),
    ("La Paz, Baja California Sur, México", -1)
]

# Función lambda para calcular la hora local
calcular_hora = lambda hora_mex, diferencia: (hora_mex + diferencia) % 24

# Hora en Ciudad de México
hora_mex = float(input("Ingrese la hora en Ciudad de México (en formato 24h): "))

# Mostrar resultados
print("Hora en diferentes ciudades del mundo")
for ciudad, diferencia in ciudades:
    hora_local = calcular_hora(hora_mex, diferencia)
    hora_int = int(hora_local)
    minutos = int((hora_local - hora_int) * 60)
    print(f"{ciudad:<40} = {hora_int:02d}:{minutos:02d}")
```

```python
def es_primo():
  while True:
    try:
      n = int(input("Ingrese un número entero: "))
      if n <= 1:
        print("El número debe ser mayor que 1.")
        continue
      Primo=True
      for i in range(2, n):
        if n % i == 0:
          Primo=False
          break
      if Primo:
        print(f"{n} es un número primo.")
        break
    except Error:
      print("Entrada inválida. Por favor, ingrese un número entero.")

es_primo()
```

```python
def analizar_numero():
  while True:
    try:
      N = int(input("Ingrese un número entero mayor que 100: "))
      if N <= 100:
        print("Error en el número. Intente nuevamente.")
        continue

      u = N % 10
      if u in [4, 7, 9]:
        print(f"{N} es divisible por 3:", N % 3 == 0)
      elif u in [1, 5]:
        print("Resultado de dividir entre 10:", N // 10)
      else:
        print("Residuo de dividir entre 3:", N % 3)
      break
    except Error:
      print("Por favor, ingrese un número válido.")

analizar_numero()
```

```python
import math

def mcd_mcm():
    try:
        a = int(input("Ingrese el primer número: "))
        b = int(input("Ingrese el segundo número: "))

        mcd = math.gcd(a, b)
        mcm = abs(a * b) // mcd

        resultados = {
            "Número A": a,
            "Número B": b,
            "MCD": mcd,
            "MCM": mcm
        }

        print("Resultados")
        for clave, valor in resultados.items():
            print(f"{clave}: {valor}")

    except Error:
        print("Error: ingrese solo números enteros válidos.")

mcd_mcm()
```

```python
def filtrar_datos_personalizado():
    print("Ingrese los elementos de su lista separados por comas (,)")
    entrada = input("Ejemplo: 25, True, 'texto', 3.14, False, 42: ")


    try:
        lista = [eval(x.strip()) for x in entrada.split(',')]
    except Exception:
        print("Error: asegúrese de escribir correctamente los valores (True, False, números o 'texto').")
        return

    enteros, flotantes, cadenas, booleanos = [], [], [], []

    for d in lista:
        if isinstance(d, bool):
            booleanos.append(d)
        elif isinstance(d, int):
            enteros.append(d)
        elif isinstance(d, float):
            flotantes.append(d)
        elif isinstance(d, str):
            cadenas.append(d)

    print("Resultados del filtrado ")
    print("Enteros:", enteros)
    print("Flotantes:", flotantes)
    print("Cadenas:", cadenas)
    print("Booleanos:", booleanos)

filtrar_datos_personalizado()
```

```python
import math

def haversine(lat1,lon1,lat2,lon2):
  R = 6371

  lat1 = math.radians(lat1)
  lon1 = math.radians(lon1)
  lat2 = math.radians(lat2)
  lon2 = math.radians(lon2)

  dlat= lat2-lat1
  dlon= lon2-lon1

  a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
  c= 2 * math.asin(math.sqrt(a))

  distancia= R * c
  return R * c

print("calcuo de las distancias")
print("introduce 3 pares de coordenadas")

for  i in range(1,4):
  print("par de coordenadas")
  lat1= float(input("latitud 1: "))
  lon1= float(input("longitud 1: "))
  lat2= float(input("latitud 2: "))
  lon2= float(input("longitud 2: "))


  distancia= haversine(lat1,lon1,lat2,lon2)
  print(f"distancia entre los puntos:{round(distancia, 2)} km")
  print("calculos para los 3 pares de coordenadas")
```

```python
def serie_fibonacci():
    while True:
        try:
            n = int(input("Ingrese la cantidad de términos de la serie de Fibonacci: "))
            if n <= 0:
                print("Error: el número debe ser mayor que 0.")
                continue

            fibo = [0, 1]  # primeros dos términos
            for i in range(2, n):
                fibo.append(fibo[i-1] + fibo[i-2])

            print("Serie de Fibonacci:")
            print(fibo[:n])  # imprime solo los n términos
            break

        except }Error:
            print("Entrada inválida. Ingrese un número entero positivo.")

serie_fibonacci()
```

```python
# Datos
Fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
           "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

Zonas = ["guerrero","oaxaca","chiapas","michoacán","guerrero",
         "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90]

# Magnitudes de str a float
magnitudes = list(map(float, magnitudes))

Zonas = list(map(lambda x: x.capitalize(), Zonas))

intensidad = []
for m in magnitudes:
    if m < 3.5:
        intensidad.append("Muy Leve")
    elif 3.5 <= m < 4.5:
        intensidad.append("Leve")
    elif 4.5 <= m < 5.5:
        intensidad.append("Moderado")
    else:
        intensidad.append("Fuerte")


conteo_por_zona = {}
for zona in Zonas:
    conteo_por_zona[zona] = conteo_por_zona.get(zona, 0) + 1


eventos = list(zip(Fechas, Zonas, magnitudes, prof_km))
eventos.sort(key=lambda x: x[2], reverse=True)


print("3 Sismos más fuertes")
for fecha, zona, mag, prof in eventos[:3]:
    print(f"{fecha} | {zona} | {mag:.1f} | {prof} km")


print("Conteo de eventos por zona")
for zona, conteo in conteo_por_zona.items():
    print(f"{zona}: {conteo} eventos")

print("Intensidad por evento ")
for i in range(len(Fechas)):
    print(f"{Fechas[i]} | {Zonas[i]} | {magnitudes[i]} | {intensidad[i]}")
```
